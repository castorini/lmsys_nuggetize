```python
import json
import pandas as pd
from datasets import load_dataset

hf_dataset = load_dataset("lmarena-ai/search-arena-v1-7k", split="test")

# filtering single-turn queries
query_ids = set()
for row in hf_dataset:
    data_all[row["question_id"]] = row
    if len(row["messages_a"]) == 2:
        query_ids.add(row["question_id"])

print(f"Total number of single-turn queries: {len(query_ids)}")
print(query_ids)

all_data = []
with open("/mnt/users/n3thakur/2025/projects/nuggetization/search-arena-v1-7k/categorization/categories.gpt-4.1.prompt.researchy.questions.temp.0.7.jsonl", "r", encoding="utf-8") as f:
    for row in f:
        data = json.loads(row)
        if data["question_id"] in query_ids:
            all_data.append(data["categories"])

len(all_data)
df = pd.DataFrame(all_data)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set the figure size
plt.rcParams["figure.figsize"] = (10,2)

# Define the columns to plot
columns = ['ambiguous', 'incompleteness', 'assumptive', 'multi-faceted', 'knowledge-intensive', 'subjective', 'reasoning-intensive', 'harmful']  # Replace with your column names

# Create a figure with multiple subplots
fig, axs = plt.subplots(1, len(columns), figsize=(16,2.3))

# Loop over the columns and create a histogram for each one
for i, col in enumerate(columns):
    data = df[col].to_list()
    sns.histplot(data, bins=10, kde=False, ax=axs[i])
    axs[i].set_title(col, fontsize=13, fontweight='bold')
    axs[i].set_xlabel('')
    axs[i].set_ylabel('')

# Layout so plots do not overlap
fig.tight_layout()

# Save the figure plot as pdf
plt.savefig("query_categories_single_turn.pdf", bbox_inches='tight')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

for language in {"English": [], "Russian": [], "Chinese": [], "German": [], "French": [], "Portuguese": []}.keys():

    # Filter the DataFrame for the current language
    all_data = []
    with open("/mnt/users/n3thakur/2025/projects/nuggetization/search-arena-v1-7k/categorization/categories.gpt-4.1.prompt.researchy.questions.temp.0.7.jsonl", "r", encoding="utf-8") as f:
        for row in f:
            data = json.loads(row)
            if data["question_id"] in inversion_qids[language]:
                all_data.append(data["categories"])

    df = pd.DataFrame(all_data)
    print(len(df))
    
    # Set the figure size
    plt.rcParams["figure.figsize"] = (10,2)

    # Define the columns to plot
    columns = ['ambiguous', 'incompleteness', 'assumptive', 'multi-faceted', 'knowledge-intensive', 'subjective', 'reasoning-intensive', 'harmful']  # Replace with your column names

    # Create a figure with multiple subplots
    fig, axs = plt.subplots(1, len(columns), figsize=(16,2.3))

    # Loop over the columns and create a histogram for each one
    for i, col in enumerate(columns):
        data = df[col].to_list()
        sns.histplot(data, bins=10, kde=False, ax=axs[i])
        axs[i].set_title(col, fontsize=13, fontweight='bold')
        axs[i].set_xlabel('')
        axs[i].set_ylabel('')
    
    axs[0].set_ylabel(f'{language}', fontsize=13, fontweight='bold')

    # Layout so plots do not overlap
    fig.tight_layout()

    # Save the figure plot as pdf
    plt.savefig(f"languages/{language.lower()}_inversion_query_categories.pdf", bbox_inches='tight')
```
